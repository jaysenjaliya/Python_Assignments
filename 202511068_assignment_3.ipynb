```python
#Question 1

def find_common_elements(list1, list2):
    common=[]
    for item in list1:
        if item in list2 and item not in common:
            common.append(item)
    return common

list1=[1,2,3,4]
list2=[3,4,5]
find_common_elements(list1,list2)
```

```python
#Question 2

my_list=[]
n=int(input("How many item you want to insert?"))

for i in range(n):
    dtype = input("enter data type(int,float,str): ")
    value = input("enter value")

    if dtype == "int":
        value = int(value)
    elif dtype == "float":
        value = float(value)
    elif dtype == "str":
        value = str(value)
    else:
        print("OOPS !! invalid datatype, storing it as a string")
        value = str(value)

    my_list.append(value)

print("List is : ",my_list)
```

```python
#Questiom-3

def count_word_frequency(text):
    words = text.lower().split()
    freq = {}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq

text = "This is a test. This test is simple."
print(count_word_frequency(text))
```

```python
#Question-4

def invert_dictionary(d):
    new_dict = {}
    for key, value in d.items():
        if value not in new_dict:
            new_dict[value] = [key]
        else:
            new_dict[value].append(key)
    return new_dict

original = {"apple": "fruit", "banana": "fruit", "carrot": "vegetable"}
print(invert_dictionary(original))
```

```python
#Question-5

employees = [
    {"ID": 1, "Name": "A", "Dept": "HR", "Salary": 30000},
    {"ID": 2, "Name": "B", "Dept": "IT", "Salary": 40000},
    {"ID": 3, "Name": "C", "Dept": "Finance", "Salary": 35000},
    {"ID": 4, "Name": "D", "Dept": "IT", "Salary": 45000},
    {"ID": 5, "Name": "E", "Dept": "HR", "Salary": 32000},
]

def add_employee(emp_id, name, dept, salary):
    employees.append({"ID": emp_id, "Name": name, "Dept": dept, "Salary": salary})

def remove_employee(emp_id):
    for emp in employees:
        if emp["ID"] == emp_id:
            employees.remove(emp)
            break

def update_salary(emp_id, new_salary):
    for emp in employees:
        if emp["ID"] == emp_id:
            emp["Salary"] = new_salary
            break

def display_employees():
    for emp in employees:
        print(emp)

print("Current Employees:")
display_employees()
```

```python
#Question-6

students = [
    {"Roll": 1, "Name": "Tom", "Marks": 85},
    {"Roll": 2, "Name": "Jerry", "Marks": 78},
]

def add_student(roll, name, marks):
    students.append({"Roll": roll, "Name": name, "Marks": marks})

def remove_student(roll):
    global students
    students = [s for s in students if s["Roll"] != roll]

def update_marks(roll, marks):
    for s in students:
        if s["Roll"] == roll:
            s["Marks"] = marks

def display_students():
    for s in students:
        print(s)


add_student(3, "Spike", 90)
update_marks(2, 88)
remove_student(1)
display_students()
```
