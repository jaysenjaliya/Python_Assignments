```python
class NoGradesError(Exception):
    pass
class InvalidGradeError(Exception):
    pass
class Student:
    def __init__(self,name):
        self.name=name
        self.grades=[]
    def __del__(self):
        print(f"Student {self.name} record created.")
    def add_grade(self,grade):
        if not (0<=grade <=100):
            raise InvalidGradeError(f" Invalid grade {grade}. Must be  between 0 and 100.")
        if len(self.grades)>=5:
            print(f"Cannot add more than 5 grades for {self.name}.")
            return 
        self.grades.append(grade)
    def average(self):
        if not self.grades:
            raise NoGradesError(f"No grades availabel for { self.name}.")
        avg=sum(self.grades)/len(self.grades)
        return round(avg,2)
    def letter_grade(self):
        avg=self.average()
        if avg>=90:
            return "A"
        elif avg>=80:
            return "B"
        elif avg>=70:
            return "C"
        elif avg>=60:
            return "D"
        else:
            return "F"
try:
    alice=Student("Alice")
    grades=[85,92,78,105]
    for g in grades:
        try:
            alice.add_grade(g)
        except InvalidGradeError as e:
            print(e)
        
    print("Alice's Grades:",alice.grades)
    print("Alice's Average:",alice.average())
    print("Alice's Letter Grade:",alice.letter_grade())
    bob=Student("Bob")
    try:
        print("Bob's Average:",bob.average())
    except NoGradesError as e:
        print(e)
finally:
    del alice
    del bob
```

```python
class BookNotFoundError(Exception):
    pass
class BookCollection:
    def __init__(self):
        self.books=[]
        self._index=0
    def add_book(self,title,author,year):
        if not (1450<= year <= 2030):
            print(f"Invalid year{year},must be between 1450 and 2030.")
            return
        if len(self.books)>=100:
            print("Collection limit reached. Cannot add more books.")
            return
        book={"title":title,"author":author,"year":year}
        self.books.append(book)
    def remove_book(self,title):
        for book in self.books:
            if book["title"]==title:
                self.books.remove(book)
                return
        raise BookNotFoundError(f"Book '{title}' not found.")
    def search_by_author(self,author):
        return [book for book in self.books if book['author'].lower()==author.lower()]
    def __iter__(self):
        self.index=0
        return self
    def __next__(self):
        while self._index<len(self.books):
            book=self.books[self._index]
            self._index+=1
            if book["year"]>2000:
                return book
        raise StopIteration
```

```python
collection=BookCollection()

collection.add_book("Python Crash Course", "Eric Matthes", 2019)
collection.add_book("Clean Code", "Robert Martin", 2008)
collection.add_book("The Pragmatic Programmer", "Hunt & Thomas", 1999)
collection.add_book("Automate the Boring Stuff", "Al Sweigart", 2015)

print("=== Iterating books after 2000 ===")
for book in collection:
    print(book)

print("\n=== Second iteration works (reset) ===")
for book in collection:
    print(book)

print("\n=== Search by author 'eric matthes' (case-insensitive)===")
print(collection.search_by_author("eric matthes"))

print("\n=== Remove book 'Clean Code' (case-sensitive match)===")
collection.remove_book("Clean Code")
for book in collection:
    print(book)

print("\n=== Try removing non-existent book===")
try:
    collection.remove_book("Nonexistent Book")
except BookNotFoundError as e:
    print(e)
```

```python
class LibraryItem:
    def __init__(self,title,item_id):
        if len(title)>200:
            raise ValueErro("Title length cannot exceed 200 characters.")
        self.title=title
        self.item_id=item_id
        self._available=True
    @property
    def status(self):
        return "Available" if self._available else "Checked out"
    def checkout(self):
        if not self._available:
            raise ItemUnavailableError(f"Item {self.item_id} already checked out.")
        self._available=False
    def return_item(self):
        if self._available:
            raise ItemNotCheckedOutError(f"Item {self.item_id} was not checked out.")
        self._available=True
```

```python
class Book(LibraryItem):
    def __init__(self,title,item_id,author,isbn):
        super().__init__(title,item_id)
        if not author or not isbn:
            raise ValueError("Book must have valid  author and ISBN.")
        self.author=author
        self.isbn=isbn
```

```python
import re
class Magazine(LibraryItem):
    def __init__(self,title,item_id,issue_number,publication_date):
        super().__init__(title,item_id)
        if issue_number<=0:
            raise ValueError("Issue number must be positive.")
        if not re.match(r"^\d{4}-(0[1-9]|1[0-2])$",publication_date):
            raise ValueError("Publication date must be in YYYY-MM format.")
        self.issue_number=issue_number
        self.publication_date=publication_date
```

```python
class Library:
    def __init__(self):
        self.items={}
    def add_item(self,item):
        if len(self.items)>=10000:
            print("Collection limit reached.")
            return
        if item.item_id in self.items:
            raise ValueError(f"Duplicate item ID{item.item_id}.")
        self.items[item.item_id]=item
    def get_item(self,item_id):
        if item_id not in self.items:
            raise ItemNotFoundError(f"Item ID {item_id} not found.")
        return self.items[item_id]
    def remove_item(self,item_id):
        if item_id not in self.items:
            raise ItemNotFoundError(f"Item ID {item_id} not found.")
        del self.items[item_id]
    def available_items(self):
        return (item for item in self.items.values() if item._available)
```

```python
library = Library()

book1 = Book(title="Design Patterns", item_id="B001", 
             author="Gang of Four", isbn="978-0201633610")
mag1 = Magazine(title="Scientific American", item_id="M001", 
                issue_number=245, publication_date="2024-03")

library.add_item(book1)
library.add_item(mag1)

print("=== Items Added ===")
for i in library.items.values():
    print(i.item_id, i.title, "-", i.status)

print("\n=== Checkout Book ===")
book1.checkout()
print(book1.title, ":", book1.status)

print("\n=== Return Book ===")
book1.return_item()
print(book1.title, ":", book1.status)

print("\n=== Iterating available items (generator) ===")
for item in library.available_items():
    print(item.item_id, item.title, "-", item.status)

print("\n=== Try invalid operations ===")
try:
    book1.return_item()  
except ItemNotCheckedOutError as e:
    print(e)

try:
    library.get_item("X999")
except ItemNotFoundError as e:
    print(e)

try:
    book1.checkout()
    book1.checkout()  
except ItemUnavailableError as e:
    print(e)
```
