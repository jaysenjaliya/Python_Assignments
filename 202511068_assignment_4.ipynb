```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def display_detail(self):
        print("Name:",self.name," and  Salary:",self.salary)
class Manager(Employee):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.dept=department
    def display_detail(self):
        print(f"Name:{self.name} ,Salary:{self.salary} ,Deparment:{self.dept} ")
e1=Employee("A",50000)
e1.display_detail()
m1=Manager("A",500000,"RC")
m1.display_detail()
```

### Question:-5

```python
class EvenNumbers:
    def __init__(self,n):
        self.n=n
        self.current=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.current >self.n:
            raise StopIteration
        else:
            even=self.current
            self.current+=2
            return even
n=int(input("Enter the last Number:"))
even_numbers=EvenNumbers(n)
for num in even_numbers:
    print(num,end=' ')
```

### Question:-6

```python
class ReverseIterator:
    def __init__(self,data):
        self.data=data
        self.index=len(data)
    def __iter__(self):
        return self
    def __next__(self):
        if self.index==0:
            raise StopIteration
        self.index-=1
        return self.data[self.index]
my_list = [1, 2, 3, 4, 5]
reverse_iter = ReverseIterator(my_list)

for item in reverse_iter:
    print(item,end=' ')
```

### Question:-7

```python
class Squares:
    def __init__(self, n):
        self.n = n
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.n:
            raise StopIteration
        result = self.current ** 2
        self.current += 1
        return result
n=int(input("Enter the number:"))
squares = Squares(n)
for square in squares:
    print(square, end=' ')
```

### Question:-8

```python
def prime_numbers(limit):
    for num in range(2,limit+1):
        is_prime=True
        for i in range(2,int(num ** 0.5)+1):
            if num%i==0:
                is_prime=False
                break
        if is_prime:
            yield num
n=int(input("Enter the number:"))
for prime in prime_numbers(n):
    print(prime,end=' ')
```

### Question:-9

```python
def fibonacci(n):
    a,b=0,1
    for _ in range(n):
        yield a
        a,b=b,a+b
n=int(input("Enter the number:"))
for num in fibonacci(n):
    print(num,end=' ')
```

```python
def make_counter():
    count = 0
    def counter():
        nonlocal count 
        count += 1
        return count

    return counter
    
counter_instance = make_counter()
print(counter_instance())  
print(counter_instance())  
print(counter_instance())  
```

### Question:-11

```python
def power_raiser(n):
    def raise_to_power(x):
        return x ** n
    return raise_to_power

square = power_raiser(2)  
cube = power_raiser(3)    

print(square(5))  
print(cube(4))    
```

### Question:-13

```python
def uppercase(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        if isinstance(result, str):  
            return result.upper()
        return result  
    return wrapper


@uppercase
def greet(name):
    return f"Hello, {name}!"

@uppercase
def get_number():
    return 123  

print(greet("Alice"))  
print(get_number())    
```
